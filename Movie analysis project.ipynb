In the 3 datasets, the first one has no missing values.
Remember one of the columns in a the box ofice movie dataset needs to be changed from an object to a float.


```python
domestic_mean = df2['domestic_gross'].mean()
domestic_median = df2['domestic_gross'].median()
domestic_mode = df2['domestic_gross'].mode()

print(f"domestic gross mean is : {domestic_mean}")
print(f"domestic gross median is : {domestic_median}")
print(f"domestic gross mode is: {domestic_mode}")
```

This shows the domestic gross data is highly skewed to the right, the mean is higher than the mode and median meaning there are outliers that actually affect the mean. We will fill the NaN values with the median since it will not disrupt the distribution of the dataset.

```python
df2['domestic_gross'].fillna(df2['domestic_gross'].median(), inplace = True)
```

```python
df2['foreign_gross'].iloc[1871:1889]
```

```python
#changing the foreign gross column from string datatype to a float.
df2['foreign_gross'] = df2['foreign_gross'].str.replace(',', '').astype('float')
```

```python
#checking if the datatype has changed.
df2['foreign_gross'].dtype
```

```python
foreign_mean = df2['foreign_gross'].mean()
foreign_median = df2['foreign_gross'].median()
foreign_mode = df2['foreign_gross'].mode()

print(f"foreign gross mean is : {foreign_mean}")
print(f"foreign gross median is: {foreign_median}")
print(f"foreign gross mode is: {foreign_mode}")
```

Ths column data shows that is also highly skewed to the right, and therefore we will replace the NaN values with the median. It has a lot of missing values but it would not be advisable to drop the whole column as it will used as part of the analysis.

```python
df2['foreign_gross'].fillna(df2['foreign_gross'].median(), inplace = True)
```

```python
#dropping the rows where there are missing values in the studio column.
df2 = df2.dropna()
```

```python
#show the genre that were produced in the years [2018,2017,2016]
movies_2016_2018 = df5[(df5['year'] == 2018) | (df5['year'] == 2017) | (df5['year'] == 2016)]
genres_by_years = movies_2016_2018['genres'].value_counts().head(10)
genres_by_years
```

```python
plt.figure(figsize=(10, 8))
genres_by_years.plot(kind = 'barh', color = 'violet')
plt.xlabel('count')
plt.ylabel('genres')
plt.title('Bar Plot of years by genres')
plt.legend()
```

```python
#genres that got the most domestic and foreign gross over the last 3 years, [2018,2017,2016]
genres_by_gross = movies_2016_2018.groupby('genres')[['domestic_gross', 'foreign_gross']].sum()
genres_by_gross['Total_Gross'] = genres_by_gross['domestic_gross'] + genres_by_gross['foreign_gross']
genres_by_gross = genres_by_gross.sort_values(by = 'Total_Gross',ascending=False).head(10)
genres_by_gross
```

```python
plt.figure(figsize=(20, 16))
genres_by_gross.plot(kind = 'barh')
plt.xlabel('values')
plt.ylabel('genres')
plt.title('Bar Plot of Values by genres')
plt.legend()
```

As seen earlier, Drama is the most watched genre, and over the years 2016-2018  it is still the most watched genre.
Also, oe the 3 years mentione dabove, Adventure, Animation,Comedy  is the genre with the higheest domestic and foreign gross.
